```python
import numpy as np
import matplotlib.pyplot as plt
```

### 독립변수 X생성

```python
x = np.random.uniform(low=0.0, high = 1.0, size = 100) #범위?
x
```

### 종속변수 y = wx+b생성

```python
#y = w*x+b
y = 7*x+5 #w값과, b값 임의로 지정
```

```python
y[:3]
```

```python
plt.scatter(x,y)
```

### 경사 하강법을 이용한 w,b학습


```python
w = np.random.uniform(low =  0.0, high = 1.0)
b = np.random.uniform(low =  0.0, high = 1.0)

learning_rate = 1.3
num_epoch = 300 #300번 학습

w,b
```

```python
for epoch in range(num_epoch):
    print(f"{epoch:3} w = {w:.5f}, b={b:.5f}")
    
    """예측한 가중치로 결과 예측"""
    y_pre=w*x+b
    
    "MSE(평균제곱오차)"
    mse = ((y_pre - y)**2).mean()
    print(f"mse : {mse:.10f}")
    
    if mse<0.0000000005 :
        break
    
    w = w-learning_rate*(2*(y_pre-y)*x).mean()
    b = b - (y_pre - y).mean()
    
print("="*50)
print(f"{epoch:3} w={w:.5f},b = {b:.5f}")
```

### Linear Model 장단점
- 장점 : 특성이 많은 데이터세트에서 성능이 좋음
- 단점 : 모델의 복잡도 제어할 방법이 없음(하이퍼파라미터 없음) ->과대적합되기쉬움
    - w의 크기가 클수록 복잡하다 -> 가중치를 조정하여 제약을 줌(w의 수치 낮춤)
    - L1규제 - w의 모든 원소에 똑같은 힘으로 규제적용 -> 특정계수들은 0이 됨. (120kg , 50kg-소멸)/ 특성선택 자동으로 이루어짐
    - L2규제 - 골고루 규제, 0에 가깝게 만듬. 
- 알파값 = 규제. 규제를 가해준다 -> 모델의 복잡도를 낮춰준다(가중치를 낮춰줌)
