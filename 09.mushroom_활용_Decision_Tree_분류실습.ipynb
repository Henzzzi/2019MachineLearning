```python
encoder = LabelEncoder()
encoder.fit(data["poisonous"])
data["poisonous"]=encoder.transform(data["poisonous"])
```

```python
encoder = LabelEncoder()
for col in data.columns:
    encoder.fit(data[col])
    data[col]=encoder.transform(data[col])
    
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=7)
```

```python
model = LogisticRegression() 
model.fit(X_train, y_train)
print("훈련점수:",model.score(X_train,y_train))
print("테스트 점수:",model.score(X_test, y_test))
```

```python
model = LogisticRegression(C=0.1) 
model.fit(X_train, y_train)
print("훈련점수:",model.score(X_train,y_train))
print("테스트 점수:",model.score(X_test, y_test))
```

```python
model = LinearSVC()
model.fit(X_train, y_train)
print("훈련점수:",model.score(X_train,y_train))
print("테스트 점수:",model.score(X_test, y_test))
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
print("훈련점수:",model.score(X_train,y_train))
print("테스트 점수:",model.score(X_test, y_test))
```

## one_hot_encoding

```python
data = pd.read_csv("./data/mushroom.csv")
x = data.iloc[:,1:]
y = data.iloc[:,0]

encoder = LabelEncoder()
y = encoder.fit_transform(y)
#y는 2차원이 될수 없기 때문에 라벨인코딩
```

```python
pd.get_dummies(["a","b","c","a","d","c"], prefix="abcd")
```

```python
x = pd.get_dummies(x, prefix=x.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, random_state=7)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = LogisticRegression() 
model.fit(X_train, y_train)
print("훈련점수:",model.score(X_train,y_train))
print("테스트 점수:",model.score(X_test, y_test))
```

```python
model = LinearSVC()
model.fit(X_train, y_train)
print("훈련점수:",model.score(X_train,y_train))
print("테스트 점수:",model.score(X_test, y_test))
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
print("훈련점수:",model.score(X_train,y_train))
print("테스트 점수:",model.score(X_test, y_test))
```

### 특성의 중요도 확인 _DecisionTree

```python
model.feature_importances_ #1에 가까울수록 중요도가 높다.
```

```python
#zip 함수 : 묶어줌 컬럼과 중요도를 ->리스트에 담아줌
list(zip(model.feature_importances_,x.columns))
```

```python
feature_list = list(zip(model.feature_importances_,x.columns))
feature_list
```

```python
#중요도에 따라 정렬
feature_list.sort()
feature_list.reverse()
feature_list
```

```python
feature = feature_list[:15]
feature
```

```python
feature_df = pd.DataFrame(feature, columns = ["score","feature"])
feature_df
```

```python
plt.bar(feature_df.feature, feature_df.score)
```

```python
plt.barh(feature_df.feature, feature_df.score) #barh 기울어짐
```
